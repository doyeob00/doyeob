```python
import tifffile as tiff
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

#   
file_path = r"C:\Users\doyeob\Desktop\3layerBP data\devided data\polar\polar data\polar_fft_91 file_1.tif"  #    .

# TIFF  
tif_data = tiff.imread(file_path)

#    
x_start, x_end = 103, 141
y_start, y_end = 165, 199
sub_region = tif_data[y_start:y_end+1, x_start:x_end+1]

# X, Y  
x = np.arange(x_start, x_end + 1)
y = np.arange(y_start, y_end + 1)
x, y = np.meshgrid(x, y)
z = sub_region

# 3D  
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

#  
ax.set_title('3D Intensity across X and Y (Range: X 103-141, Y 165-199)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Intensity')

plt.show()
```
