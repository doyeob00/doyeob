```python
import tifffile as tiff
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 데이터 파일 경로
file_path = r"C:\Users\doyeob\Desktop\3layerBP data\devided data\polar\polar data\polar_fft_91 file_1.tif"  # 주어진 파일 경로로 변경하세요.

# TIFF 파일 읽기
tif_data = tiff.imread(file_path)

# 특정 범위 데이터 추출
x_start, x_end = 103, 141
y_start, y_end = 165, 199
sub_region = tif_data[y_start:y_end+1, x_start:x_end+1]

# X, Y 좌표 생성
x = np.arange(x_start, x_end + 1)
y = np.arange(y_start, y_end + 1)
x, y = np.meshgrid(x, y)
z = sub_region

# 3D 플롯 생성
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# 축 설정
ax.set_title('3D Intensity across X and Y (Range: X 103-141, Y 165-199)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Intensity')

plt.show()
```
