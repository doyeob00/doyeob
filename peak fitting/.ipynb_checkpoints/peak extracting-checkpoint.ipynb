```python
import os
import re
import numpy as np
import tifffile as tiff
import pandas as pd
```

```python
##################### 101A1

# 데이터 파일 경로
folder_path = r"C:\Users\doyeob\Desktop\3layerBP data\devided data\polar\polar data"
output_excel_path = r"C:\Users\doyeob\Desktop\3layerBP data\devided data\polar\101A-1 infromation.xlsx"
delta_x, delta_y = 1, 1

# 결과를 저장할 리스트 초기화
coordinates = []
angular_profiles = []
radial_profiles = []

# 폴더 내의 모든 TIFF 파일에 대해 반복
for filename in os.listdir(folder_path):
    if filename.endswith(".tif"):
        file_path = os.path.join(folder_path, filename)
        
        # TIFF 파일 읽기
        tif_data = tiff.imread(file_path)
        
        # 특정 범위 데이터 추출
        x_start, x_end = 116, 138
        y_start, y_end = 172, 199
        sub_region = tif_data[y_start:y_end+1, x_start:x_end+1]
        
        # 최대 픽셀 값 및 좌표 찾기
        max_pixel_value = np.max(sub_region)
        max_coords = np.unravel_index(np.argmax(sub_region), sub_region.shape)
        max_y, max_x = max_coords[0] + y_start, max_coords[1] + x_start  # 전체 이미지 기준 좌표
        
        # 파일명에서 마지막 숫자 추출
        file_number = int(re.findall(r'\d+', filename)[-1])
        
        # 결과 저장 (좌표 정보)
        coordinates.append([file_number, max_x, max_y, max_pixel_value])
        
        # 선 프로파일 생성
        y_profile = np.mean(tif_data[:, max_x-delta_x:max_x+delta_x+1], axis=1)
        x_profile = np.mean(tif_data[max_y-delta_y:max_y+delta_y+1, :], axis=0)
        
        # 각 프로파일에 파일 번호와 함께 저장
        angular_profiles.append([file_number] + y_profile.tolist())
        radial_profiles.append([file_number] + x_profile.tolist())

# 데이터프레임 생성
coordinates_df = pd.DataFrame(coordinates, columns=['File Number', 'Max X', 'Max Y', 'Max Intensity'])
angular_df = pd.DataFrame(angular_profiles)
radial_df = pd.DataFrame(radial_profiles)

# 첫 번째 열을 "File Number"로 이름 변경
angular_df.rename(columns={0: 'File Number'}, inplace=True)
radial_df.rename(columns={0: 'File Number'}, inplace=True)

# 파일 번호를 기준으로 정렬
coordinates_df = coordinates_df.sort_values(by='File Number').reset_index(drop=True)
angular_df = angular_df.sort_values(by='File Number').reset_index(drop=True)
radial_df = radial_df.sort_values(by='File Number').reset_index(drop=True)

# 행과 열을 반전 (transpose)
angular_df = angular_df.transpose()
radial_df = radial_df.transpose()

# 엑셀 파일로 저장
with pd.ExcelWriter(output_excel_path) as writer:
    coordinates_df.to_excel(writer, sheet_name='Coordinates', index=False)
    angular_df.to_excel(writer, sheet_name='Angular', header=False, index=False)
    radial_df.to_excel(writer, sheet_name='Radial', header=False, index=False)

print('저장완료')
```
