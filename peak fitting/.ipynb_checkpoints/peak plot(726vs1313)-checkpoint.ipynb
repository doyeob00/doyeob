```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 가우시안 함수 정의
def gaussian(x, a, b, c):
    return a * np.exp(-((x - b) ** 2) / (2 * c ** 2))

# 726th 데이터 피크 파라미터
params_726 = [8.67340555e+05, 1.05074788e+00, 1.58205476e-01,
              1.57847568e+06, 1.76178069e+00, 4.50345925e-01,
              3.25404797e+06, 2.34878019e+00, 2.86951627e-01,
              8.03098411e+05, 2.92530527e+00, 2.01627427e-01,
              4.71381470e+05, 3.41306224e+00, 4.27939590e-01]

# 1313th 데이터 피크 파라미터
params_1313 = [1.08286277e+06, 1.06640085e+00, 1.43669434e-01,
               1.27141924e+06, 1.56792881e+00, 4.49883936e-01,
               3.81743880e+06, 2.31229269e+00, 3.39250318e-01,
               1.08707252e+06, 2.92943196e+00, 2.26995717e-01,
               9.73111465e+05, 3.47147481e+00, 4.06568347e-01]

# x 데이터 범위 설정
x_data = np.linspace(0, 5, 500)

# 각 피크 개별 가우시안 계산 및 플롯
fig, axes = plt.subplots(5, 1, figsize=(10, 20), sharex=True)

for i in range(5):
    a_726, b_726, c_726 = params_726[3*i:3*i+3]
    a_1313, b_1313, c_1313 = params_1313[3*i:3*i+3]
    
    peak_726 = gaussian(x_data, a_726, b_726, c_726)
    peak_1313 = gaussian(x_data, a_1313, b_1313, c_1313)
    
    axes[i].plot(x_data, peak_726, label=f'726th Peak {i+1}', linestyle='--')
    axes[i].plot(x_data, peak_1313, label=f'1313th Peak {i+1}', linestyle=':')
    axes[i].axvline(x=b_726, color='blue', linestyle='--', linewidth=0.5)
    axes[i].axvline(x=b_1313, color='orange', linestyle=':', linewidth=0.5)
    axes[i].legend()
    axes[i].set_ylabel('Intensity')
    axes[i].set_xticks(np.arange(0, 5.5, 0.5))

axes[-1].set_xlabel('x')
plt.suptitle('Comparison of Gaussian Peaks: 726th vs 1313th')
plt.show()
```
