```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 엑셀 데이터 읽기
file_path = r"D:\Users\doyeob\Desktop\3layerBP data\gaussian fitting\input data\726th data.xlsx"
data = pd.read_excel(file_path)

# 데이터 시각화
x_data = data.iloc[:, 0]
y_data = data.iloc[:, 1]

plt.figure(figsize=(10, 6))
plt.plot(x_data, y_data, label='Data')
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Data Plot')
plt.grid(True)
plt.show()

# 가우시안 함수 정의
def gaussian(x, amp, cen, wid):
    return amp * np.exp(-(x - cen)**2 / (2 * wid**2))

# 첫 번째 피크 피팅
initial_guess_1 = [max(y_data[:len(y_data)//2]), 1, 0.1]
popt1, _ = curve_fit(gaussian, x_data, y_data, p0=initial_guess_1, bounds=([0, 0.5, 0], [np.inf, 1.5, np.inf]))
fitted_peak1 = gaussian(x_data, *popt1)

# 두 번째 피크 피팅
initial_guess_2 = [max(y_data[len(y_data)//2:]), 2.4, 0.1]
popt2, _ = curve_fit(gaussian, x_data, y_data, p0=initial_guess_2, bounds=([0, 2, 0], [np.inf, 3, np.inf]))
fitted_peak2 = gaussian(x_data, *popt2)

# 피팅 결과 시각화
plt.figure(figsize=(10, 6))
plt.plot(x_data, y_data, label='Data')
plt.plot(x_data, fitted_peak1, label='Fitted Peak 1', linestyle='--')
plt.plot(x_data, fitted_peak2, label='Fitted Peak 2', linestyle='--')
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Gaussian Fitting')
plt.legend()
plt.grid(True)
plt.show()

# 피팅 결과 출력
print("First Peak: Amplitude = {:.2f}, Center = {:.2f}, Width = {:.2f}".format(*popt1))
print("Second Peak: Amplitude = {:.2f}, Center = {:.2f}, Width = {:.2f}".format(*popt2))
```
